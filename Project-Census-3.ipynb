```python

# Age Pyramid

#Generates list of ages for male and female
age_male = df['Age'][df['Gender'] == 'Male'].tolist()
age_female = df['Age'][df['Gender'] == 'Female'].tolist()

def get_age(lists, st_age, end_age):
    return sum (1 for i in lists if st_age <= i <= end_age)

AgeClass = ['95+','89-94','84-89','80-84','75-79','70-74','65-69','60-64','55-59', '50-54', '45-49', '40-44', '35-39', '30-34', 
            '25-29', '20-24', '15-19', '10-14','5-9', '0-4']

male_agecount=[]
female_agecount=[]
for lar in AgeClass:
    start=0
    end=0
    if '-' in lar:
        jk = lar.split('-')
        start = int(jk[0])
        end = int(jk[1])
    else:        
        start = 100
        end = 120
    male_agecount.append(-1* get_age(age_male, start, end))
    female_agecount.append(get_age(age_female, start, end)) 

#_________________________________________________________________________________________________

age_p = pd.DataFrame({'Age': AgeClass, 
                      'Male':male_agecount, 
                      'Female': female_agecount})

plt.figure(figsize= (14, 10))
sns.set_style("white")
bar_plot = sns.barplot(x='Male', y='Age', data=age_p, order=AgeClass, palette = 'RdYlGn')
bar_plot = sns.barplot(x='Female', y='Age', data=age_p, order=AgeClass, palette = "gist_earth_r")

bar_plot.set(xlabel="Population (hundreds)", ylabel="Age-Group", title = "Population Pyramid")
#plt.show()

for p in bar_plot.patches:
    
    # Get X and Y placement of label from rect.
    x_value = p.get_width()
    y_value = p.get_y() + p.get_height() / 2

    # Number of points between bar and label. Change to your liking.
    space = 5
    # Vertical alignment for positive values
    ha = 'left'

    # If value of bar is negative: Place label left of bar
    if x_value < 0:
        # Invert space to place label to the left
        space *= -1
        # Horizontally align label at right
        ha = 'right'

    # Use X value as label and format number with one decimal place
    label = int(x_value)

    # Create annotation
    plt.annotate(
        label,                      # Use `label` as label
        (x_value, y_value),         # Place label at end of the bar
        xytext=(space, 0),          # Horizontally shift label by `space`
        textcoords="offset points", # Interpret `xytext` as offset in points
        va='center',                # Vertically center label
        ha=ha, fontsize=10.5)         # Horizontally align label differently for
                                    # positive and negative values.
    
(bar_plot.get_figure()).savefig('./pyramid.png')
```
